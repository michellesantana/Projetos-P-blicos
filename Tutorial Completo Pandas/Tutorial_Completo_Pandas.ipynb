## Adicionando uma coluna:

```
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed'] # coluna incluida igual ao somatório de outras colunas

df.head(5)
```

### Outra maineira de adicionar coluna:

```
df['Total'] = df.iloc[:, 4:10].sum(axis=1) #somando coluna 4 até a coluna 10  #loc/iloc/axis????
df
```

## Eliminando uma coluna:

```
df= df.drop(columns=['Total']) #Eliminando uma coluna
df
```

## Movendo posição das colunas:

```
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df

# Exercício: descobrir índice de cada coluna
```

## Filtrando dados (Baseada em múltiplas condições):

### Filtrando uma coluna:

```
df.loc[df['Type 1'] == 'Grass'] # Filtrando uma coluna
```

### Filtrando duas coluna (E ou OU):

```
#para usar OU então usar '|' exemplo:

df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')] # Filtrando uma opção ou outra
```

```
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')] # Filtrando duas colunas diferentes

#& = and

## E filtrar duas opções na mesma coluna?
```

```
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)] # Filtrando três colunas diferentes
```

## Salvando os dados filtrados em um novo arquivo:

```
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)] # Filtrando três colunas diferentes
new_df.to_csv('filtered.csv') #salvando os dados filtrados em um novo arquivo 

#OBS: quando esses dados filtrados são salvos, permanecem os índices antigos
```

### Salvando e resetando o índice (index):

```
new_df = new_df.reset_index(drop=True) #Drop = True (retira o antigo índice)
new_df

#or

#new_df.reset_index(drop=True, inplace=True) #inplace=True ???
```

## Filtrando um termo em texto:

```
df.loc[df['Name'].str.contains('Mega')] #Filtrando na coluna 'Name' termos com 'Mega'
```

```
df.loc[~df['Name'].str.contains('Mega')] #Filtrando na coluna 'Name' termos sem 'Mega' (usamos o tio (~))
```

```
import re
df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)] #Filtrando na coluna 'Type 1' 'Fire' or 'Grass, usar '~' para o contrário
```

```
df.loc[df['Type 1'].str.contains('Fire|Grass', flags=re.I, regex=True)] #Filtrando na coluna 'Type 1' 'Fire' or 'Grass, usar '~' para o contrário
```

```
df.loc[df['Name'].str.contains('pi[a-z]*', flags=re.I, regex=True)] #Filtrando na coluna 'Name' nomes que contenham 'pi' e a sequência é qualquer letra entre A e Z

#Para que serve o '*'??
```

```
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)] #Filtrando na coluna 'Name' nomes que começam com 'pi' (usar: '^')
```

## Mudanças Condicionais:

```
df
```

```
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer' # Mudando a palavra 'Fire' por 'Flame' na coluna 'Type 1'
df
```

```
df.loc[df['Type 1'] == 'Flamer', 'Type 1'] = 'Fire' # Mudando a palavra 'Fire' por 'Flame' na coluna 'Type 1' - Voltando a mudança realizada na célula acima
df
```

```
df.loc[df['Type 1'] == 'Fire', 'Legendary'] = True #Racional: se tem 'Fire' na coluna 'Type 1', então na coluna 'Legendary' colocar 'True'
df
```

### Lendo arquivo de dados novamente:

```
df1 = pd.read_csv('modified.csv') #Lendo arquivo de dados modificado anteriormente
df1
```

```
#Racional: Se o 'Total' é maior que 500, então na coluna 'Legendary' e 'Generation' colocar: 'TEST VALUE'
df1.loc[df1['Total'] > 500, ['Generation', 'Legendary']] = 'TEST VALUE'
df1
```

```
#Racional: Se o 'Total' é maior que 500, então na coluna 'Legendary' e 'Generation' colocar: 'TEST 1' e 'TEST 2' respectivamente:
df1.loc[df1['Total'] > 500, ['Generation', 'Legendary']] = ['TEST 1', 'TEST 2']
df1
```

## Estatísticas Agregadas (Groupby):

```
df1
```

```
df1.groupby(['Type 1']).mean() #Média de todas as colunas em relação ao 'Type 1'
```

```
df1.groupby(['Type 1']).mean().sort_values('Defense', ascending=False) #Média de todas as colunas em relação ao 'Type 1' com a coluna 'Defense' em ordem decrescente
```

```
f1.groupby(['Type 1']).mean().sort_values(['Defense', 'Attack'], ascending=False) #Média de todas as colunas em relação ao 'Type 1' com a coluna 'Defense' e 'Attack' em ordem decrescente
```

```
#df1.groupby(['Type 1']).count() #Contagem. Porém, para evitar colunas vazias, vamos criar uma coluna 'count':

df1['count'] = 1 #Criando coluna 
df1.groupby(['Type 1']).count()['count']
```

```
df1.groupby(['Type 1','Type 2']).count()['count'] #Contagem por 'Type 1' e 'Type 2'
```
